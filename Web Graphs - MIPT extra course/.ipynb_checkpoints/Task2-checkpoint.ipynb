# Исследование модели Боллобаша-Риордана

```python
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np
import pylab
from collections import Counter

%matplotlib inline
```

### Задание 1
Построить граф $G_{m}^{n}$ в модели Боллобаша-Риордана (n = 10000, m = 2).

```python
def generateSimpleBollobasRiordanGraph(n):
    G = nx.MultiDiGraph()
    
    G.add_node(0)
    G.add_edge(0, 0)
    
    repeated_nodes = [0, 0] #все вершины; каждая вершина повторяется в количестве, равном её степени
    
    while G.number_of_nodes() < n:
        new_node = G.number_of_nodes()
        
        G.add_node(new_node)
        repeated_nodes.append(new_node) #для повзможной петли
        
        destination = random.choice(repeated_nodes) #выбираем конец ребра случайным образом пропрционально степени
        
        repeated_nodes.append(destination)
        G.add_edge(new_node, destination)
    
    return G
```

```python
def generateBollobasRiordanGraph(n, m):
    G1mn = generateSimpleBollobasRiordanGraph(n * m)
    G = nx.MultiDiGraph()
    
    #объединяем вершины по группам {0, ..., m-1}, {m, ..., 2*m - 1}, ..., {m(n-1), ..., mn - 1}
    for u, v in G1mn.edges():
        u_new, v_new = u // m, v // m
        G.add_edge(u_new, v_new)
    
    return G
```

```python
G = generateBollobasRiordanGraph(10000, 2)
```

```python
print(G.number_of_nodes())
print(G.number_of_edges())
```

### Задание 2
Построить графики распределения степеней вершин графа в модели Боллобаша-Риодана и теоретической оценки.

Теоретическая оценка количества вершин степени d в графе $G_{m}^{n}$ равна
$$\frac{2 \cdot m \cdot (m + 1) \cdot n}{d (d + 1) (d + 2)}$$

```python
def degree_estimation(d, m, n):
    return 2 * m * (m + 1) * n / d / (d + 1) / (d + 2)
```

```python
N = 10000
M = 2
G = generateBollobasRiordanGraph(N, M)
```

```python
#рисуем рельное распределение
degrees = Counter(dict(nx.degree(G)).values())
sorted_degree_values = sorted(degrees.keys())
counts = [degrees[d] for d in sorted_degree_values]

plt.loglog(sorted_degree_values, counts, ls='None', marker='o', markersize=8, color='c')

#рисуем теоретическое распределение
degree_values = np.arange(1, max(dict(G.degree()).values()) + 1, 0.2)[:800]
theoretical_values = [degree_estimation(d, M, N) for d in degree_values]

plt.loglog(degree_values, theoretical_values, color='r', ls='--', lw=3)

#названия осей
plt.title("Degree Distribution", fontsize=20)
plt.ylabel("nodes (log)", fontsize=20)
plt.xlabel("degree (log)", fontsize=20)

pylab.rcParams['figure.figsize'] = 10, 10
plt.show()
```

Вывод: реальное распределение соответсвует теоретическому.

### Задание 3
Исследовать устойчивость гигантской компоненты в графах $G_{m}^{n}$ в модели Боллобаша-Риордана. Для m = 3 построить график зависимости доли вершин в самой большой компоненте связности при удалении вершин с вероятностью p = 0.3 от n = {100, 200, ..., 10000}.

```python
M = 3
p = 0.3
```

```python
nodes_counts, fractions = [], []

for N in range(100, 10100, 100):
    G = generateBollobasRiordanGraph(N, M)
    
    #удаляем вершины с вероятностью p
    for v in list(G.nodes()):
        if random.random() < p:
            G.remove_node(v)
    
    largest_connected_component_size = len(next(nx.connected_components(nx.MultiGraph(G))))
    fractions.append(largest_connected_component_size / G.number_of_nodes())
    nodes_counts.append(N)
```

```python
pylab.rcParams['figure.figsize'] = 8, 8
plt.ylim(0.8, 1)

plt.plot(nodes_counts, fractions)

plt.title('m = %d' % M, fontsize=20)
plt.ylabel('доля вершин в Гиг. компоненте', fontsize=20)
plt.xlabel('N', fontsize=20)

plt.show()
```

```python
#среднее значение
sum(fractions) / len(fractions)
```

Вывод: графы в модели Боллобаша-Риордана обладают устойчивостью гигантской компоненты.

### Задание 4
Сгенерировать граф $H_{a, m}^{n}$ в модели Бакли-Остгуса на n = 5000 вершинах при m = 2, a = 0.3. Построить графики распределения степеней вершин и теоретической оценки.

```python
def generateSimpleBuckleyOsthusGraph(a, n):
    G = nx.MultiDiGraph()
    
    G.add_node(0)
    G.add_edge(0, 0)
    
    while G.number_of_nodes() < n:
        new_node = G.number_of_nodes()
        G.add_node(new_node)
        
        #считаем вероятности для концов рёбер
        probabilities = [
            (a - 1 + nx.degree(G, i)) / ((a + 1.0) * new_node + a)
            for i in range(new_node)
        ]
        
        #добавляем вероятность петли
        probabilities.append(a / ((a + 1.0) * new_node + a))
        
        destination = np.random.choice(G.nodes(), 1, p=probabilities)[0]
        G.add_edge(new_node, destination)
        
    return G      
```

```python
def generateBuckleyOsthusGraph(a, n, m):
    G1mn = generateSimpleBuckleyOsthusGraph(a, m * n)
    G = nx.MultiDiGraph()
    
    #объединяем вершины по группам {0, ..., m-1}, {m, ..., 2*m - 1}, ..., {m(n-1), ..., mn - 1}
    for u, v in G1mn.edges():
        u_new, v_new = u // m, v // m
        G.add_edge(u_new, v_new)
    
    return G
```

По теореме Гречникова доля вершин степени d в $H_{a, m}^{n}$ равна
$$\beta_{a, m, d} = \frac{B(d - m + ma, a + 2)}{B(ma, a + 1)} = \Theta (d^{- 2 - a})$$

```python
from scipy import special
def beta(a, m, d):
    return special.beta(d - m + m * a, a + 2) / special.beta(m * a, a + 1)
```

```python
n = 5000
a = 0.3
m = 2
G = generateBuckleyOsthusGraph(a, n, m)
```

```python
#рисуем рельное распределение
degrees = Counter(dict(nx.degree(G)).values())
sorted_degree_values = sorted(degrees.keys())
counts = [degrees[d] for d in sorted_degree_values]

plt.loglog(sorted_degree_values, counts, ls='None', marker='o', markersize=8, color='c')

#рисуем теоретическое распределение
degree_values = np.arange(1, max(dict(G.degree()).values()) + 1, 0.2)[4:800]
theoretical_values = [beta(a, m, d) * n for d in degree_values]

plt.loglog(degree_values, theoretical_values, color='r', ls='--', lw=3)

#названия осей
plt.title("Degree Distribution", fontsize=20)
plt.ylabel("nodes (log)", fontsize=20)
plt.xlabel("degree (log)", fontsize=20)

pylab.rcParams['figure.figsize'] = 10, 10
plt.show()
```

Вывод: распределение степенй вершин в модели Бакли-Остгуса соответствует теоретической оценке.
